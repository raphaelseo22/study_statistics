## Homework 1

You are an Analyst working for a high-end clothes design boutique.

The company is developing a new line of clothes for very tall people. Your team is analyzing the viability of the project from a sales perspective and your manager has asked you to assist with some input variables to help test the financial forecast.

You need to three distributions:
- A normal distribution of 1,000 observation for heights of men in US
- A normal distribution of 1,000 observation for heights of women in US

Alse for eacj of the two populations you have been asked to identify the minimun height of 2.2% of the tallest people in the population.

In the US men's heights have a mean of 69.1 inches (175.5cm) and standard deviation 2.9 inches (7.4cm), while females heights have a mean of 63.7 inches (161.8cm) and standard deviation 2.7 inches (6.9cm)

```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
m_heights_mean = 69.1
m_heights_std = 2.9
w_heights_mean = 63.7
w_heights_std = 2.7
```

### Men heights

```python
print(m_heights_mean + 2 * m_heights_std)
```

```python
m_rv = norm(loc = m_heights_mean, scale = m_heights_std)
```

```python
m_rv.ppf(0.978)
```

```python
m_random_v1 = m_rv.rvs(1000, random_state=0).round(1)
m_random_v2 = m_rv.rvs(1000, random_state=1).round(1)
m_random_v3 = m_rv.rvs(1000, random_state=2).round(1)
```

```python
plt.hist(m_random_v1, density=True, bins="auto", histtype="stepfilled", alpha=0.2)
plt.plot(np.sort(m_random_v1), norm.pdf(np.sort(m_random_v1), m_heights_mean, m_heights_std))
plt.show
```

```python
plt.hist(m_random_v2, density=True, bins="auto", histtype="stepfilled", alpha=0.2)
plt.plot(np.sort(m_random_v2), norm.pdf(np.sort(m_random_v2), m_heights_mean, m_heights_std))
plt.show
```

```python
plt.hist(m_random_v3, density=True, bins="auto", histtype="stepfilled", alpha=0.2)
plt.plot(np.sort(m_random_v3), norm.pdf(np.sort(m_random_v3), m_heights_mean, m_heights_std))
plt.show
```
